**Creating New Feature using "Cabin" Feature**

1. Cabin feature consists of deck/num//side , where deck is deck loacation, num is deck_number and side can be P for port or S for Starboard.
2. Separate all these 3 values from cabin & create three new features Cabin_Deck, Cabin_Number & Cabin_Side.
3. Cabin feature is having NaN values so to avoid error while splitting we have to replace it in such a way that we can split those NaN Values in all three new features respectively.

```python
def cabin_new_features(df):
    df['Cabin'].fillna("np.nan/np.nan/np/nan").astype(str)
    
    df['Cabin_Deck'] = df['Cabin'].apply(lambda x: x.split('/')[0] if isinstance(x, str) else np.nan)
    df['Cabin_Number'] = df['Cabin'].apply(lambda x: x.split('/')[1] if isinstance(x, str) else np.nan)
    df['Cabin_Side'] = df['Cabin'].apply(lambda x: x.split('/')[2] if isinstance(x, str) else np.nan)
    
    #Replacing string nan values to numpy nan values.
    cols = ["Cabin_Deck","Cabin_Number","Cabin_Side"]
    df[cols]=df[cols].replace("np.nan",np.nan)
    
     #Filling Missing Values in new features created.
    df["Cabin_Deck"].fillna(df["Cabin_Deck"].mode()[0])
    df["Cabin_Side"].fillna(df["Cabin_Side"].mode()[0])
   
    df["Cabin_Number"] = pd.to_numeric(df["Cabin_Number"], errors="coerce")
    df["Cabin_Number"] = df["Cabin_Number"].fillna(df["Cabin_Number"].median())
```

```python
cabin_new_features(train_df)
cabin_new_features(test_df)
```

# Data Pre-Processing

1. Missing values

```python
print(f'Missing values in train set: ({np.round(100*train_df.isna().sum()/len(train_df),1)}%)')
print(f'Missing values in test set: ({np.round(100*test_df.isna().sum()/len(test_df),1)}%)')
```

```python
data = [train_df, test_df]
for dataset in data:
    numerical_features = list(dataset.columns[(dataset.dtypes =='float64')| (dataset.dtypes =='int')])
    categorical = list(dataset.columns[dataset.dtypes =='object'])
```

```python
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy="most_frequent")     ##To fill Categorical Features.
imputer2 = SimpleImputer(strategy="median")            ##To fill numeircal features.
```

```python
def fill_miss(df):
    df[numerical_features] = imputer2.fit_transform(df[numerical_features])
    df[categorical]= imputer1.fit_transform(df[categorical])
    
fill_miss(train_df)
fill_miss(test_df)
```

```python
print("Missing numbers left in train_df is:",train_df.isnull().sum().sum())
print("Missing numbers left in test_df is:",test_df.isnull().sum().sum())
```

```python
train_df[categorical].head(10)
```

# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
# Initialize the encoder
encoder = OneHotEncoder()

encoder.fit(train_df[nominal_categorical])
categories=[]
for i,feature in enumerate(nominal_categorical):
    categories.extend([f'{feature}:{value}' for value in encoder.categories_[i]])
train_df.loc[:,categories] = encoder.transform(train_df[nominal_categorical]).toarray().astype(int)
# Remove original Cabin column from the dataset since it's not needed anymore.
train_df= train_df.drop(nominal_categorical, axis=1)

encoder.fit(test_df[nominal_categorical])
categories=[]
for i,feature in enumerate(nominal_categorical):
    categories.extend([f'{feature}:{value}' for value in encoder.categories_[i]])
test_df.loc[:,categories] = encoder.transform(test_df[nominal_categorical]).toarray().astype(int)
test_df = test_df.drop(nominal_categorical, axis=1)
```

**Download libraries**

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score, recall_score, confusion_matrix
```

```python
result_train = []
result_test = []
```

```python
def model_prediction(model):
    df=[]
    model.fit(X_train1, y_train1.values.ravel())
    X_train_pred1 = model.predict(X_train1)
    accuracy = accuracy_score(y_train1,X_train_pred1)*100
    f1 = f1_score(y_train1,X_train_pred1)
    recall = recall_score(y_train1,X_train_pred1)
    result_train.append(accuracy)
    df.append(["Train",accuracy,f1,recall])
    
    X_test_pred1 = model.predict(X_test1)
    accuracy= accuracy_score(y_test1,X_test_pred1)*100
    f1 = f1_score(y_test1,X_test_pred1)
    recall = recall_score(y_test1,X_test_pred1)
    result_test.append(accuracy)
    df.append(["Test",accuracy,f1,recall])
    
    print(f"Confusion Matrix of {model} model is:")
    cm = confusion_matrix(y_test1,X_test_pred1)
    plt.figure(figsize=(8,4))
    sns.heatmap(cm,annot=True,fmt="g",cmap="summer")
    plt.show()
    
    return pd.DataFrame(columns=["Group","accuracy", "f1","recall"], data =df)
```

# Hyper Parameter Tunning of RandomForest Model

```python
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import GridSearchCV, HalvingGridSearchCV
```

```python
Model_up_Forest = RandomForestClassifier()
parameters_Forest = {'n_estimators': [100,200,300],
                     'max_depth': [5,7,10],
                     'min_samples_split': [4,6],
                     'min_samples_leaf':[4,6]}
```

```python
grid_search_forest = GridSearchCV(Model_up_Forest,parameters_Forest,cv=5, n_jobs=-1)
grid_search_forest.fit(X_train,y_train.values.ravel())
```

```python
grid_search_forest.best_score_
best_parameters_forest = grid_search_forest.best_params_
```

```python
best_parameters_forest
```

```python
Model_up_Forest = RandomForestClassifier(**best_parameters_forest)
Model_up_Forest.fit(X_train,y_train.values.ravel())
X_test_pred_Forest = Model_up_Forest.predict(X_test)
accuracy_score(y_test,X_test_pred_Forest)*100
```

```python
gsh = HalvingGridSearchCV(Model_up_Forest, parameters_Forest, factor=3,  cv=5)
gsh.fit(X_train,y_train.values.ravel())
```

```python
gsh.best_params_
```

```python
Model_up_Forest1 = RandomForestClassifier(**gsh.best_params_)
Model_up_Forest1.fit(X_train,y_train.values.ravel())
X_test_pred_Forest1 = Model_up_Forest1.predict(X_test)
accuracy_score(y_test,X_test_pred_Forest1)*100
```
